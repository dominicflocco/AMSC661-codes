```python
from distmesh import * 
import numpy as np 
import scipy as sc
from scipy.sparse import csr_matrix
from scipy.sparse import diags
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Generates mesh on domain $\Omega = [-1,1]^2$.

```python
h = 0.04 # mesh step

# Fixed Points 
pfix = np.array([[1,-1], [1,1], [-1,1], [-1,-1]])
bbox = np.array([-1,1,-1,1])

# Distance function 
def fd(p):
    return drectangle(p,-1,1,-1,1)
pts, tri = distmesh2D(fd, huniform, h, bbox, pfix)

# Save mesh to matlab workspace file
sc.io.savemat('mesh.mat', {'pts': pts, 'tri': tri})
```

```python
# Plot mesh 
plt.rcParams.update({
        "font.family": "Times", 
        "text.usetex": True,
        "font.size": 16
    })
plt.triplot(pts[:,0],pts[:,1],tri,linewidth=0.1)
axes=plt.gca()
axes.set_aspect(1)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Triangulated Mesh on Domain $\Omega = [-1,1]^2$')
```
